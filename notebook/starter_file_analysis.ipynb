```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shamimhasan8/python-code-bug-and-fix-pairs")

print("Path to dataset files:", path)
```

```python
import pandas as pd

data = pd.read_csv("data/code_bug_fix_pairs.csv", header=0)
data
```

```python
snippet_df = (
    pd.concat([data["buggy_code"], data["fixed_code"]], ignore_index=True).dropna().to_frame(name="code snippet").sample(frac=1, random_state=42).reset_index(drop=True)
)
snippet_df
```

```python
snippet_df = snippet_df.reset_index().rename(columns={"index": "id"})
snippet_df
```

```python
snippet_df.info()
```

```python
snippet_df.to_csv("code_snippets.csv", index=False)
```
